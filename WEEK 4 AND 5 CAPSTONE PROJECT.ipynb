```python
address = 'Chicago, IL'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Chicago city are {}, {}.'.format(latitude, longitude))
```

```python
Chicago_map= folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, description in zip(dfgeom['LATITUDE'], dfgeom['LONGITUDE'], dfgeom['DESCRIPTION']):
    label = '{}, {}'.format(dfgeom, description)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Chicago_map)  
    
Chicago_map
```
